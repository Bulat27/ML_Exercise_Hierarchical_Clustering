# Loading the dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/Bulat27/ML_Exercise_Hierarchical_Clustering/master/Wholesale%20customers%20data.csv'
df = pd.read_csv(url)
```

```python
df.head(5)
```

# Prepocessing the data

```python
from sklearn.preprocessing import normalize

df_scaled = normalize(df)
df_scaled = pd.DataFrame(df_scaled, columns=df.columns)
```

```python
df_scaled.head()
```

# Plotting the dendogram

```python
import scipy.cluster.hierarchy as shc
 
plt.figure(figsize=(11, 8))   
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(df_scaled, method='ward'))
```

```python
plt.figure(figsize=(11, 8))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(df_scaled, method='ward'))
plt.axhline(y=6, color='r')
```

# Applying the agglomerative clustering

```python
from sklearn.cluster import AgglomerativeClustering

agglomerative_clustering = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')  
agglomerative_clustering.fit_predict(df_scaled)
```

# Visualization of the clusters

```python
plt.figure(figsize=(11, 8))  
plt.scatter(df_scaled['Milk'], df_scaled['Grocery'], c=agglomerative_clustering.labels_) 
```
